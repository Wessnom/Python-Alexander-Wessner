## 1. Pythagorean theorem (*)

### a)   A right angled triangle has the catheti: a = 3 and b = 4 length units. Compute the hypothenuse of the triangle. (*)
### b)   A right angled triangle has hypothenuse c = 7.0 and a cathetus a = 5.0 length units. Compute the other cathetus and round to one decimal. (*)


**Solution a)**

```python
import math

a = 3.0 # Length of side 'a'
b = 4.0 # Length of side 'b'

# Calculate the hypotenuse using Pythagorean theorem
c = math.sqrt(a**2 + b**2)

print(f"The hypotenuse (side 'c') is {c:.2} length units.")
```

**Solution b)**

```python
import math

c = 7.0 # Length of side 'c'
a = 5.0 # Length of side 'a'

# Calculate the cathetus using Pythagorean theorem
b = math.sqrt(c**2 - a**2)

print(f"The length of cathetus 'b' is {b:.2} length units.")
```

# 2. Classification accuracy (*)
A machine learning algorithm has been trained to predict whether or not it would rain the next day. Out of 365 predictions, it got 300 correct. Compute the accuracy of this model.

```python
total_predictions = 365
correct_predictions = 300

accuracy = (correct_predictions / total_predictions)

print(f"The accuracy of the model is {accuracy:.0%}.")
```

# 3. Classification accuracy (*)


```python
TP = 2 # True Positive
FP = 2 # False Positive
FN = 11 # False Negative
TN = 985 # True Negative

# Calculate accuracy
accuracy = (TP + TN) / (TP + FP + FN + TN)

print(f"Accuracy: {accuracy:.3f}")
```

# 4. Line (*)
Compute the slope ***k*** and the constant term ***m*** of this line using the points **A : (4,4) and B : (0,1)**

```python
# Given points
x1, y1 = 4, 4
x2, y2 = 0, 1

# Calculate the slope, k
k = (y2 - y1) / (x2 - x1)

# Calculate the constant term (m) aka intersection with y-axis using one of the given points. 
m = y1 - k * x1

print(f"The slope (k) is {k}.")
print(f"The constant term (m) is {m}.")
```
