## 1. Dice rolls (*)

Simulate 10 dice rolls and **append** the rolls to a list or use **list comprehension**. 

&nbsp; a) &nbsp; sort the list in **ascending** order (*)

&nbsp; b) &nbsp; sort the list in **descending** order (*)

&nbsp; c) &nbsp; find the **maximum** and **minimum** value in the lsit (*)

````python
# A1)
import random

dice_rolls = []

for _ in range(10): # _ throwaway variable, won't be used in loop
    roll = random.randint(1, 6) # Rolls a six sided die
    dice_rolls.append(roll) # Append the results to the list

print(dice_rolls)

````

````python
# A2)
# Or with list comprehension:
import random

dice_rolls = [random.randint(1, 6) for _ in range(10)]

print(dice_rolls)

# Sorting in ascending order:

dice_rolls.sort()
print(dice_rolls)
````

````python
# B) 

dice_rolls.sort(reverse=True)
print(dice_rolls)
````

````python
# C) 

max_value = max(dice_rolls)
print(max_value)

min_value = min(dice_rolls)
print(min_value)
````

## Food menu (*)

Create a;

&nbsp; a) &nbsp; **list** with the following elements: "vegetarisk lasagne", "spaghetti", "fisk", "grönsakssoppa", "pannkakor". (*)

&nbsp; b) &nbsp; **list** with the weekdays (*) 

&nbsp; c) &nbsp; create a food menu with each day corresponding to each food item and print it out. (*)

````python
food_menu = ["vegetarisk lasagne", "spaghetti", "fisk", "grönsakssoppa", "pannkakor"]
print(food_menu)

weekdays = ["måndag", "tisdag", "onsdag", "torsdag", "fredag"]

print("Veckomeny")
for day, food in zip(weekdays, food_menu):
    print(f"{day}: {food}")
````

## 3. Squares (*)

&nbsp; a) &nbsp; Use **list comprehension** to create a list of squares from -10 to 10 (*)

&nbsp; a) &nbsp; Plot this list using **matplotlib**. (*)

````python
import matplotlib.pyplot as plt

list_of_squares = [x ** 2 for x in range(-10, 11)] # 11 non inclusive
print(list_of_squares)

x_values = [x for x in range(-10, 11)]

plt.figure(figsize=(10, 6))

plt.plot(x_values, list_of_squares, marker='o', linestyle='-', color='b')

plt.xlabel('X values')
plt.ylabel('Squares')
plt.title('Squares of numbers from -10 to 10')
plt.grid(True)
plt.show
````
