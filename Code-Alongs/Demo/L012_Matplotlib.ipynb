# Matplotlib

```python
import matplotlib.pyplot as plt

x = list(range(-10, 11)) # Not inclusive so meaning -10 to 10
y = [value**2 for value in x] # Using list comprehension, y = x**2 
y2 = [value**3 for value in x]

print(f"{x = }")
print(f"{y = }")
print(f"{y2 = }")

plt.plot(x, y, label="x**2")
plt.plot(x, y2,label="x**3")
plt.plot(x, y, color='red', linestyle='dashed', marker='o')
plt.plot(x, y2, color='blue', linestyle='dashed', marker='o')
plt.legend()
plt.grid()
```

## Sub plots

```python
x = list(range(-10, 11)) # Not inclusive so meaning -10 to 10
y = [value**2 for value in x] # Using list comprehension, y = x**2 
y2 = [value**3 for value in x]

fig, axs = plt.subplots(1, 2) # 1 rad med 2 kolumner
axs[0].plot(x, y)
axs[1].plot(x, y2)
```
